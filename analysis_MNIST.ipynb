```python
%load_ext autoreload
%autoreload 2

import pickle
import time
import tensorflow as tf
import numpy as np

from analysis import *

%matplotlib inline
```

```python
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("MNIST_data/", one_hot=False)

Xtr = mnist.train.images.reshape(-1, 28, 28, 1)
Ytr = mnist.train.labels.astype(float)

Xtt = mnist.test.images.reshape(-1, 28, 28, 1)
Ytt = mnist.test.labels.astype(float)

val_set = {'X': Xtt[:500], 'Y': Ytt[:500]}
C2 = np.mean([np.sqrt(np.sum(np.square(i))) for i in Xtr])
gamma = 0.04*C2
eps_wrm = 1./(2*gamma)

eps_wrm = 1.3 # Note the above eps_wrm is 1.356, but we'll use 1.3 as used in the paper's code
```

```python
from spectral_adversarial_regularization.models import elunet as model

arch = model.elunet
save_dir = os.path.join('save_weights', 'mnist', 'wrm')
train_network(Xtr, Ytr, val_set, arch, save_dir, eps=eps_wrm, adv='wrm', retrain=False)
```

```python
# Estimate eps for the other samples using the perturbation achieved for each test sample
save_dir = os.path.join('save_weights', 'mnist', 'wrm')
arch = model.elunet

Xtr_adv_wrm = ad.build_graph_and_gen_adv_examples(Xtr, arch, save_dir, beta=1, num_channels=1,
                                                  load_epoch=25, method=ad.wrm, eps=eps_wrm)
    
eps = np.sqrt(np.mean([np.sum(np.square(i)) for i in Xtr_adv_wrm-Xtr])) 

eps = 0.05*C2 # Note: the above eps/C2 is 0.06, but we'll use 0.05 as seen in Figure 2a of the paper
```

```python
arch = model.elunet
methods = ['erm', 'fgm', 'pgm']

for method in methods:
    save_dir = os.path.join('save_weights', 'mnist', method)
    train_network(Xtr, Ytr, val_set, arch, save_dir, eps=eps, adv=method, retrain=False)
```

```python
# Evaluate robustness of networks
eps_list = np.linspace(0, 0.25, 9)*C2
methods = ['erm', 'fgm', 'pgm', 'wrm']
curves_file = 'save_weights/mnist/adv_attack_curves_loadepoch25.pickle'

if os.path.isfile(curves_file):
    adv_results = pickle.load(file(curves_file, 'rb'))
else:
    adv_results = {}
    for method in methods:
        save_dir = os.path.join('save_weights_final', 'mnist', method)
        s_norm = dl_utils.get_overall_sn(save_dir, arch, num_channels=1, load_epoch=25)
        acc, adv_accs = get_adv_acc_curve(Xtt, Ytt, save_dir, arch, eps_list, load_epoch=25)
        adv_results[method] = (acc, adv_accs, s_norm)
    pickle.dump(adv_results, file(curves_file, 'wb'))
        
plot_acc_curves(adv_results, eps_list/C2)
```

```python
# Taking a closer look at some images after adversarial perturbation
eps = 0.3*C2

n = 5
np.random.seed(0)
ind_samp = np.random.choice(range(len(Xtt)), n, replace=False)
Xtt_samp = Xtt[ind_samp]

adv_examples = {}
for method in methods:
    save_dir = os.path.join('save_weights', 'mnist', method)
    X_adv = ad.build_graph_and_gen_adv_examples(Xtt_samp, arch, save_dir, num_channels=1,
                                                method=ad.wrm, eps=eps, order=2, load_epoch=25)
    Y_adv = dl_utils.build_graph_and_predict(X_adv, save_dir, arch, beta=1., num_channels=1, load_epoch=25)
    adv_examples[method] = (X_adv, Y_adv)
    
for i in range(n):
    plt.figure(figsize=(16, 3))
    plt.subplot(1, 5, 1)
    plt.imshow(Xtt_samp[i].reshape(28, 28), cmap='gray')
    plt.title('original image')
    for ii, method in enumerate(methods):
        plt.subplot(1, 5, ii+2)
        plt.imshow(adv_examples[method][0][i].reshape(28, 28), cmap='gray')
        plt.axis('off')
        plt.title('%s adv (labeled %s)'%(method, adv_examples[method][1][i]))
    plt.tight_layout()
    plt.show()
```

```python
# Experiment to show how WRM networks fit on random labels easily change their labelings
# based on small perturbations to the network

arch = model.elunet
eps_list = np.linspace(0, 0.25, 9)*C2
curves_file = 'save_weights/mnist/wrm_test_perturbation_curves.pickle'

cargs = {'order':2, 'method':ad.pgm, 'beta': 1., 'load_epoch':200}

if os.path.exists(curves_file):
    adv_results = pickle.load(file(curves_file, 'rb'))
    
else:
    adv_reults = {}
    
    save_dir = 'save_weights/mnist/wrm_randlabels/'
    
    Ytrhat = dl_utils.build_graph_and_predict(Xtr, save_dir, arch, num_channels=1, load_epoch=200)
    _, adv_results['Rand labels (refit on training set)'] = \
        get_adv_acc_curve(Xtr, Ytrhat, save_dir, arch, eps_list, **cargs)

    Ytthat = dl_utils.build_graph_and_predict(Xtt, save_dir, arch, num_channels=1, load_epoch=200)
    _, adv_results['Rand labels (refit on test set)'] = \
        get_adv_acc_curve(Xtt, Ytthat, save_dir, arch, eps_list, **cargs)

    save_dir = 'save_weights/mnist/wrm/'

    Ytrhat = dl_utils.build_graph_and_predict(Xtr, save_dir, arch, num_channels=1, load_epoch=200)
    _, adv_results['True labels (refit on training set)'] = \
        get_adv_acc_curve(Xtr, Ytrhat, save_dir, arch, eps_list, **cargs)

    Ytthat = dl_utils.build_graph_and_predict(Xtt, save_dir, arch, num_channels=1, load_epoch=200)
    _, adv_results['True labels (refit on test set)'] = \
        get_adv_acc_curve(Xtt, Ytthat, save_dir, arch, eps_list, **cargs)
    
    pickle.dump(adv_results, file(curves_file, 'wb'))

plot_acc_curves(adv_results, eps_list/C2, logy=False)
```

```python
# Repeat above with spectrally-normalized WRM network

arch = model.elunet_sn
beta = 2.0
eps_list = np.linspace(0, 0.25, 9)*C2
curves_file = 'save_weights/mnist/wrm_beta%s_test_perturbation_curves.pickle'%(beta)

cargs = {'order':2, 'method':ad.pgm, 'beta': beta, 'load_epoch':200}

if os.path.exists(curves_file):
    adv_results = pickle.load(file(curves_file, 'rb'))
    
else:
    adv_results = {}
    
    save_dir = 'save_weights/mnist/wrm_beta%s_randlabels/'%(beta)
    
    Ytrhat = dl_utils.build_graph_and_predict(Xtr, save_dir, arch, num_channels=1, load_epoch=200, beta=beta)
    _, adv_results['Rand labels (refit on training set)'] = \
        get_adv_acc_curve(Xtr, Ytrhat, save_dir, arch, eps_list, **cargs)

    Ytthat = dl_utils.build_graph_and_predict(Xtt, save_dir, arch, num_channels=1, load_epoch=200, beta=beta)
    _, adv_results['Rand labels (refit on test set)'] = \
        get_adv_acc_curve(Xtt, Ytthat, save_dir, arch, eps_list, **cargs)

    save_dir = 'save_weights/mnist/wrm_beta%s/'%(beta)

    Ytrhat = dl_utils.build_graph_and_predict(Xtr, save_dir, arch, num_channels=1, load_epoch=200, beta=beta)
    _, adv_results['True labels (refit on training set)'] = \
        get_adv_acc_curve(Xtr, Ytrhat, save_dir, arch, eps_list, **cargs)

    Ytthat = dl_utils.build_graph_and_predict(Xtt, save_dir, arch, num_channels=1, load_epoch=200, beta=beta)
    _, adv_results['True labels (refit on test set)'] = \
        get_adv_acc_curve(Xtt, Ytthat, save_dir, arch, eps_list, **cargs)
    
    pickle.dump(adv_results, file(curves_file, 'wb'))

plot_acc_curves(adv_results, eps_list/C2, logy=False, title='WRM with SN = %s (PGM attacks)'%(beta))
```

```python
adv = 'wrm'
curves_file = 'save_weights/mnist/%s_perturbation_curves.pickle'%(adv)

if os.path.exists(curves_file):
    adv_results = pickle.load(file(curves_file, 'rb'))

else:
    adv_results = {}
    
    for f in sorted(os.listdir('save_weights/mnist/')):
        if adv in f and 'backup' not in f and 'pickle' not in f:
            save_dir = os.path.join('save_weights', 'mnist', f)

            if 'beta' in f:
                arch = model.elunet_sn
                beta = float(f.split('beta')[1].split('_')[0])
                key_base = r'%s $\beta$ = %s'%(adv.upper(), beta)
            else:
                arch = model.elunet
                beta = 0
                key_base = '%s'%(adv.upper())

            if 'rand' in f:
                key_base += ' Rand'
            else:
                key_base += ' True'

            cargs = {'beta': beta, 'load_epoch':200, 'num_channels':1}

            Ytrhat = dl_utils.build_graph_and_predict(Xtr, save_dir, arch, **cargs)
            _, adv_results['%s (train_refit)'%(key_base)] = \
                get_adv_acc_curve(Xtr, Ytrhat, save_dir, arch, eps_list, method=ad.pgm, **cargs)

            Ytthat = dl_utils.build_graph_and_predict(Xtt, save_dir, arch, **cargs)
            _, adv_results['%s (test_refit)'%(key_base)] = \
                get_adv_acc_curve(Xtt, Ytthat, save_dir, arch, eps_list, method=ad.pgm, **cargs)
                
    pickle.dump(adv_results, file(curves_file, 'wb'))
```
